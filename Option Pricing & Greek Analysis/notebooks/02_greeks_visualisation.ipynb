```python
import numpy as np
import matplotlib.pyplot as plt
S_values = np.linspace(50, 150, 100)
K = 100
T = 1
r = 0.05
sigma = 0.2
delta_call = [param(S0, K, T, r, sigma) for S0 in S_values]
plt.plot(S_values,delta_call)
plt.xlabel("Spot Price (S0)")
plt.ylabel("Delta Call") 
plt.title("Delta Call vs Spot Price")
plt.grid(True)
plt.show()

```
