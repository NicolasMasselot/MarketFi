```python
import sys
sys.path.append("../src")

from pricing import call_price, put_price
```

```python
import matplotlib.pyplot as plt
import numpy as np
#visualisation 1 = Call price vs Strike
S0 = 100
T = 1
r = 0.05
sigma = 0.2
K_values = np.linspace(50,150,100)
y = [call_price(S0,K,T,r,sigma) for K in K_values]
plt.plot(K_values,y)
plt.xlabel("Strike")
plt.ylabel("Call Price") 
plt.title("Call Price vs Strike")
plt.grid(True)
plt.show()

```

```python
#visualisation 2 = Call price vs Volatility
S0 = 88
K = 102
T = 1
r = 0.05
sigma_values = np.linspace(0.1,0.6,100)
y = [call_price(S0,K,T,r,sigma) for sigma in sigma_values]
plt.plot(sigma_values,y)
plt.xlabel("Volatility")
plt.ylabel("Call Price") 
plt.title("Call Price vs Volatility")
plt.grid(True)
plt.show()
```
